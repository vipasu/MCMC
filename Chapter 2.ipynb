## Problem 1
a) Define mean, variance, skew, and kurtosis.

b) Compute these statistics for a uniform distribution

c) Generate K random numbers and plot the estimates as a function of log$_2$K for K = $4^n$ for $n=1,...,10$

### a)

Mean - Assuming the arithmetic mean, simply the sum of a set of values divided by the number of values. Also known as the expected value of a distribution.

$$Mean(x) = \frac{1}{n}\Sigma_{i=1}^n x_i == E[x] == \bar{x}$$

Variance - Expectation of the squared deviation of a random variable from it's mean. Can also be calculated by taking the second moment and subtracting the square of the first moment.

$$Var(x) = E[(x-\bar{x})^2]$$

Skew - Measures the asymmetry of a distribution about its mean. Positve skew means that there is larger tail to the right.

$$Skew(x) = E\left[\left(\frac{x-\bar{x}}{\sigma}\right)^3\right]$$ where $\sigma$ is the standard deviation of $x$

Kurtosis - Corresponds to a scaled version of the fourth moment of a distribution. A high kurtosis value means more mass lies in the tails. 

$$Kurt(X) = \frac{E[(x-\bar{x})^4]}{\sigma^4}$$

### b)
For the uniform distribution on [0,1], we have

$$\bar{x} = \int_0^1 x dx = \frac{1}{2}$$

$$Var(x) = \int_0^1 \left(x-\frac{1}{2}\right)^2 dx = \frac{1}{12} = \sigma^2$$

$$Skew(x) = \int_0^1 \left(\frac{x - \frac{1}{2}}{\frac{1}{\sqrt{12}}}\right)^3 dx = 0$$

$$Kurt{x} = \int_0^1 \frac{\left(x-\frac{1}{2}\right)^4}{\sigma^4} dx = 1.8$$


The interval is from 0 to 1, so the normalization factor of $\frac{1}{1-0} = 1$ is omitted.

One may also note that the Skew should be 0 just by the symmetry of the distribution

### c)

```julia
using PyPlot 
```

```julia
for n in collect(1:10)
    k = 4^n
    x = rand(k)
    mu = mean(x)
    sigma = std(x)
    var = sigma^2
    skew = mean(((x-mu)/sigma).^3)
    kurt = mean((x-mu).^4/sigma.^4)
    scatter(log2(k), mu, color="blue")
    scatter(log2(k), var, c="green")
    scatter(log2(k), skew, c="red")
    scatter(log2(k), kurt, c="black")
end
xlabel(L"log$_2$ K")
ylabel("Value")
axhline(0.5, color="b")
axhline(1./12, color="g")
axhline(0, color="r")
axhline(1.8, color="k")
legend(["Mean", "Var", "Skew", "Kurt"])
```
